```python
from googletrans import Translator

translator = Translator()

# Dịch một câu
text = "Xin chào, bạn có khỏe không?"
result = await translator.translate(text, src='vi', dest='en')
print(f"{result.origin} -> {result.text}")

# Dịch nhiều câu
texts = ['Xin chào', 'Tôi yêu Python', 'Hôm nay trời đẹp']
translations = await translator.translate(texts, src='vi', dest='en')
for t in translations:
    print(f"{t.origin} -> {t.text}")

# Phát hiện ngôn ngữ
detected = await translator.detect('Bonjour tout le monde')
print(f"Ngôn ngữ: {detected.lang}, độ tin cậy: {detected.confidence}")
```
